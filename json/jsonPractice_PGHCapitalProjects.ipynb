```python
with open('spamspam.txt', 'w') as f:
...     print('This will be written to somedir/spamspam.txt', file=f)# parses a JSON file listing capital projects and their locations in Pittsburgh
import csv

# define constants
areas = []
LOG_FILE = 'malformedProjects.txt'
AREA_TYPES_FILE = 'projectAreaTypes.txt'

def printProject(proj):
    print("***PROJECT PROFILE***")
    for key in proj:
        print(key, end=': ')
        print(proj[key])

def assembleAreaList(proj):
    with open(AREA_TYPES_FILE, 'a') as areaFile:
        try:
            dex = areas.index(proj['area'])
            print("Found area at %i" % dex)
            print("Skipping area: %s" % proj['area'])
        except:
            areas.append(proj['area'])
            print("Adding area: %s" % proj['area'])
            areaFile.write(str(proj['area']))
            areaFile.write('\n')
        
def logMalformedProject(proj):
    with open(LOG_FILE, 'a') as outfile:
        outfile.write(str(proj['id']))
    
```

```python
import json
f = open('cgcapitalprojects_img.geojson', 'r')
# alltext = f.readlines()
# print(alltext)
res = json.load(f)
# print(res)
for p in res['features']:
    printProject(p['properties'])
    assembleAreaList(p['properties'])
    if p['properties']['area'] == '':
        logMalformedProject(p['properties'])
print(areas)
        
```
