```python
# Screen scraping demo: CCAC online course catalog search results
import requests



def getResultString(searchTerm):
    url = 'https://coursesections.ccac.edu/Pages/CoursesResults.aspx?k=%s' % (searchTerm)
    print(url)
    res = requests.get(url)
    
    return res.text
```

```python
page = getResultString('cit')
# print(page)
idx = page.find('Microsoft.SharePoint.Client.Search.Query.ResultTableCollection')
print(page)
print(str(page[idx:idx+100]))
print(len(page))
```

```python
import urllib
from bs4 import BeautifulSoup
p = 'https://www.goodreads.com/search?utf8=%E2%9C%93&query=python'

req = urllib.request.Request(p)
with urllib.request.urlopen(req) as response:
    the_page = response.read()
soup = BeautifulSoup(the_page, 'html.parser')
for e in soup.find_all("a", title=True):
    print(e)
```
